```python
from PIL import Image, ImageFilter 
import matplotlib.pyplot as plt

# Opens a image in RGB mode

im = Image.open(r"c:\Users\mahes\Downloads\light-dawn-landscape-1533483.jpg")

# Blurring the image

im1 = im.filter(ImageFilter.BoxBlur(15))

# Shows the image in image viewer

plt.imshow(im1)

plt.title("Blurred Image")

plt.axis('on')
```
