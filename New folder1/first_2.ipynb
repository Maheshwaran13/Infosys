```python
import librosa

import numpy as np

import matplotlib.pyplot as plt

#Load the audio file

audio, sr = librosa.load('/content/VideoAudio.mp4')

#Calculate the amplitude

amplitude= np.abs(audio)

#Calculate the frequency

n_fft = 2048

hop_length= 512

stft= np.abs(librosa.stft(audio, n_fft=n_fft, hop_length=hop_length))

Freqs= np.mean(stft, axis=1)

times =librosa.frames_to_time(np.arange(stft.shape[1]), sr=sr, hop_length=hop_length) #short time fo

#Plot the amplitude over time

plt.figure(figsize=(12, 6))

plt.plot(np.arange(len(audio)) / sr, amplitude)

plt.xlabel('Time (s)')

plt.ylabel('Amplitude')

plt.title('Amplitude over Time')

plt.show()

#Plot the frequency over time

plt.figure(figsize=(12, 6))

plt.imshow(librosa.amplitude_to_db(stft, ref=np.max), cmap='Accent', origin="lower", aspect="auto")

plt.xlabel('Time (s)')

plt.ylabel('Frequency Bin')

plt.title("Frequency over Time")

plt.colorbar(format='%+2.of dB')

plt.show()
```

Splitting the dataset into train and test sets is one of the most important steps in machine learning.

 Why do we split the dataset into Train and Test sets?

Because we want to check how well our model performs on new, unseen data.

 Think of it like studying for an exam

Training data = what you study

Testing data = final exam questions

If you test yourself using the same questions you studied, you will get a high scorebut that doesnt mean you truly understand the subject.

Similarly

If a model is tested on the same data it was trained on, it may look good but will fail on real-world data.
