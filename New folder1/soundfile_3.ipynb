```python
import numpy as np      #used for numeric operations
import soundfile as sf   #used to save audio data into a WAV file
from IPython.display import Audio, display  #allows audio playback inside jupyter/ collab

#SETTING

duration = 3

sample_rate = 16000 #16 kHz audio  howmany sample per second

frequency =440 #44 tone (can change) -the pitch of the tone, and can able to change it

#**** GENERATE SINE WAVE

t = np.linspace(0, duration, int(sample_rate * duration),endpoint=False)

audio_data= 0.5 * np.sin(2* np.pi * frequency * t) # this line generete sine wave sound # What is a Sine Wave?

                                                    #A sine wave is a smooth, periodic wave described by the mathematical function:

#SAVE AUDIO FILE

output_file= "generated_audio.wav"

sf.write(output_file, audio_data, sample_rate)  #this writes the numppy array into a WAV file using the specified sample rate.

#PRINT OUTPUT

print("Audio file created successfully!")

print("Saved as:", output_file)

#PLAY AUDIO IN COLAB.

display(Audio(output_file, autoplay=True))
```

# audio file through meta

```python
# audio file through meta


import numpy as np
import soundfile as sf
from IPython.display import Audio, display
from mutagen.wave import WAVE
from mutagen.id3 import TextFrame

#===== SETTINGS =====

duration=3  # seconds
sample_rate = 16000 # 16 kHz audio
frequency = 440  #Sine tone frequency

#===== GENERATE SINE WAVE ======

t= np.linspace(0, duration, int (sample_rate * duration), endpoint=False)
audio_data = 0.5 * np.sin(2 * np.pi * frequency * t)

#===== SAVE AUDIO FILE =====

output_file = "generated_audio_with_info.wav"
sf.write(output_file, audio_data, sample_rate)

#===== ADD METADATA =====

metadata = WAVE (output_file)


#Use TextFrame for metadata values

metadata ["INAM"] = TextFrame(encoding=3, text=["Sample Tone Audio"]) # INAME = title or name of the audio

metadata["IART"] = TextFrame (encoding=3, text=["Your Name"]) #IART = artist

metadata["ICMT"] = TextFrame(encoding=3, text="This is a generated audio tone with embedded metadata.") #CMT= comment

metadata["ICRD"] = TextFrame(encoding=3, text=["2025-82-14"])

metadata.save()

#PLAY AUDIO

print("Audio file created with metadatal\n")

display (Audio(output_file, autoplay=False))

#READ AND PRINT METADATA

print(" === Embedded Metadata in WAV File ===")

for key, value in metadata.items():
    print(f"(key): {value.text[0] if hasattr(value, 'text') else value}") # Access text attribute for TextFrame

```

```python
from gtts import gTTS

from IPython.display import Audio, display
import soundfile as sf
import numpy as np

#TEXT TO SPEECH INPUT

text = "Hello! This is an aiiya moodra super da dai generated audio message created in Google Colab."

#GENERATE AUDIO FROM TEXT

tts= gTTS(text=text, lang='en')
tts.save("text_audio.mp3")

#Convert MP3 to WAV (optional)

#Load MP3 using audio libraries

import librosa

audio_data, sr= librosa.load("text_audio.mp3", sr=16000)
sf.write("text_audio.wav", audio_data, sr)

#PLAY AUDIO

print("Text converted to speech and saved as 'text_audio.wav'") 
display(Audio("text_audio.wav", autoplay=False))
```
