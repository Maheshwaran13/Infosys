TO CONVER TEXT TO AUDIO

```python
with open('sample.txt','w') as f:
  f.write("This text is going to convert into audio")
```

```python
from gtts import gTTS
from IPython.display import display,Audio
text = open("sample.txt").read()
tts = gTTS(text=text,lang = 'en')
tts.save("sample_audio.mp3")

display(Audio('sample_audio.mp3',autoplay=False))
```

convert mp3 to wav

```python
import shutil, subprocess, sys
from pathlib import Path

# Use a Path object to avoid backslash-escape problems and handle spaces
mp3_path = Path('sample_audio.mp3')
# If the relative file doesn't exist, try the notebook directory explicitly
if not mp3_path.exists():
    mp3_path = Path(r'C:\internship infosys\New folder2') / 'sample_audio.mp3'

# Prefer system ffmpeg if available; otherwise use a Python fallback
if shutil.which('ffmpeg'):
    subprocess.run(["ffmpeg", "-y", "-i", str(mp3_path), "-ar", "16000", "-ac", "1", "audio.wav"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    print("Converted to audio.wav (16k mono) using system ffmpeg")
else:
    print("System ffmpeg not found. Installing Python fallback (pydub + imageio-ffmpeg)...")
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', 'pydub', 'imageio-ffmpeg'])
    from pydub import AudioSegment
    import imageio_ffmpeg as _ffmpeg
    AudioSegment.converter = _ffmpeg.get_ffmpeg_exe()
    audio = AudioSegment.from_file(str(mp3_path))
    audio = audio.set_frame_rate(16000).set_channels(1)
    audio.export('audio.wav', format='wav')
    print("Converted to audio.wav (16k mono) using pydub + imageio-ffmpeg")
```
